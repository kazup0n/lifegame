```python
def count_livings(y, x):
    livings = 0
    # x=0 => -1 はダメ
    # x=1 => 両方OK
    # x=2 => +1はダメ

    # y=0 => -1 はダメ
    # y=1 => 両方OK
    # y=2 => +1はダメ


    # y = -1
    if x != 0 and y != 0  and cells[y-1][x-1] == 1:
        livings = livings + 1
    if y != 0 and cells[y-1][x] == 1:
        livings = livings + 1
    if x != 2 and y != 0 and cells[y-1][x+1] == 1:
        livings = livings + 1

    # y = 0
    
    if x != 0 and cells[y][x-1] == 1:
        livings = livings + 1
    if x != 2 and cells[y][x+1] == 1:
        livings = livings + 1

    # y = +1
    if x != 0 and y != 2 and cells[y+1][x-1] == 1:
        livings = livings + 1
    if y != 2 and cells[y+1][x] == 1:
        livings = livings + 1
    if x != 2 and y != 2  and cells[y+1][x+1] == 1:
        livings = livings + 1
    
    return livings

def calc_next_state(y, x):

    count = count_livings(y,x)
    next_state = cells[y][x]

    # 誕生 死んでいるセルに隣接する生きたセルがちょうど3つあれば、次の世代が誕生する。
    if cells[y][x] == 0 and count == 3:
        next_state = 1

    # 生存 生きているセルに隣接する生きたセルが2つか3つならば、次の世代でも生存する。
    if cells[y][x] == 1 and (count == 2 or count == 3):
        next_state = 1

    # 過疎 生きているセルに隣接する生きたセルが1つ以下ならば、過疎により死滅する。
    if cells[y][x] == 1 and count <= 1:
        next_state = 0

    # 過密 生きているセルに隣接する生きたセルが4つ以上ならば、過密により死滅する。
    if cells[y][x] == 1 and count >= 4:
        next_state = 0

    return next_state

def print_row(row):
    for cell in row:
        print(cell_disp[cell], end='')
    print('')
        
def print_cells():
    print_row(cells[0])
    print_row(cells[1])
    print_row(cells[2])
    print('')


# セルの状態を表示する文字
cell_disp = {}
cell_disp[0] = '💤'
cell_disp[1] = '😻'



# 初期状態
# 1=> live, 0 => dead
cells = [[1, 0, 1],
         [1, 1, 0],
         [1, 0, 0]
        ]
# 計算する最大世代数
generations = 20
# 現在の世代
count = 0

while count < generations:
    print_cells()
    # 次の状態を計算
    new_cells = [[0,0,0], [0,0,0], [0,0,0]]

    new_cells[0][0] = calc_next_state(0, 0)
    new_cells[0][1] = calc_next_state(0, 1)
    new_cells[0][2] = calc_next_state(0, 2)

    new_cells[1][0] = calc_next_state(1, 0)
    new_cells[1][1] = calc_next_state(1, 1)
    new_cells[1][2] = calc_next_state(1, 2)

    new_cells[2][0] = calc_next_state(2, 0)
    new_cells[2][1] = calc_next_state(2, 1)
    new_cells[2][2] = calc_next_state(2, 2)
    cells = new_cells

    count = count + 1
```
